```python
import pandas as pd
import matplotlib.pyplot as plt

#  
df = pd.read_csv("Most Streamed Spotify Songs 2023.csv", encoding='latin1')

#       
print(df.head())

#  1:      10  
top_songs = df.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_songs['track_name'], top_songs['streams'], color='skyblue')
plt.xlabel(' ')
plt.title('10    Spotify  2023 ')
plt.gca().invert_yaxis()  #      
plt.show()

#  2:       danceability
plt.figure(figsize=(8, 8))
plt.pie(df['danceability_%'].head(10), labels=df['track_name'].head(10), autopct='%1.1f%%')
plt.title('  danceability     2023 ')
plt.show()

#  3:           2023 
# (,   "Track Name"   0)
track_name = df.loc[0, 'track_name']
track_data = df[df['track_name'] == track_name]
plt.figure(figsize=(10, 6))
plt.plot(track_data['released_day'], track_data['streams'], marker='o')
plt.xlabel('')
plt.ylabel(' ')
plt.title(f'    "{track_name}"  2023 ')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

#  4:     10  
top_artists = df['artist(s)_name'].value_counts().nlargest(10)

#   
plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='salmon')
plt.xlabel('')
plt.ylabel(' ')
plt.title('10    Spotify  2023    ')
plt.xticks(rotation=45)
plt.show()
#  5:      
#     (,       )
tracks_data = df.head(3)
plt.figure(figsize=(10, 6))
plt.barh(tracks_data['track_name'], tracks_data['streams'], color=['skyblue', 'salmon', 'green'])
plt.xlabel(' ')
plt.title('       2023 ')
plt.gca().invert_yaxis()
plt.show()
```
