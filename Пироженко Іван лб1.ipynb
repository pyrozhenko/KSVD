```python
import pandas as pd
import matplotlib.pyplot as plt

# Завантажити дані
df = pd.read_csv("Most Streamed Spotify Songs 2023.csv", encoding='latin1')

# Вивести перші кілька рядків для перегляду структури
print(df.head())

# Завдання 1: Побудувати горизонтальну стовпчикову діаграму для 10 найпопулярніших пісень
top_songs = df.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_songs['track_name'], top_songs['streams'], color='skyblue')
plt.xlabel('Кількість прослуховувань')
plt.title('10 найпопулярніших пісень на Spotify у 2023 році')
plt.gca().invert_yaxis()  # Щоб пісні були в порядку спадання
plt.show()

# Завдання 2: Побудувати кругову діаграму для розподілу за danceability
plt.figure(figsize=(8, 8))
plt.pie(df['danceability_%'].head(10), labels=df['track_name'].head(10), autopct='%1.1f%%')
plt.title('Розподіл за danceability серед найпопулярніших пісень у 2023 році')
plt.show()

# Завдання 3: Побудувати лінійний графік для змін кількості прослуховувань певного треку протягом 2023 року
# (припустимо, що трек "Track Name" має індекс 0)
track_name = df.loc[0, 'track_name']
track_data = df[df['track_name'] == track_name]
plt.figure(figsize=(10, 6))
plt.plot(track_data['released_day'], track_data['streams'], marker='o')
plt.xlabel('Дата')
plt.ylabel('Кількість прослуховувань')
plt.title(f'Зміни кількості прослуховувань пісні "{track_name}" у 2023 році')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Завдання 4: Побудувати стовпчикову діаграму для 10 найпопулярніших артистів
top_artists = df['artist(s)_name'].value_counts().nlargest(10)

# Побудувати стовпчикову діаграму
plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='salmon')
plt.xlabel('Артист')
plt.ylabel('Кількість пісень')
plt.title('10 найпопулярніших артистів на Spotify у 2023 році за кількістю пісень')
plt.xticks(rotation=45)
plt.show()
# Завдання 5: Порівняти кількість прослуховувань трьох конкретних треків
# Виберемо три конкретні треки (припустимо, що це перші три треки з даних)
tracks_data = df.head(3)
plt.figure(figsize=(10, 6))
plt.barh(tracks_data['track_name'], tracks_data['streams'], color=['skyblue', 'salmon', 'green'])
plt.xlabel('Кількість прослуховувань')
plt.title('Порівняння кількості прослуховувань трьох конкретних треків у 2023 році')
plt.gca().invert_yaxis()
plt.show()
```
